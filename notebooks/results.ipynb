```python
def get_results(file, dictionary, iterations= iterations):
  for i , j in enumerate(dictionary.keys()):
    #print(i,j)
    for k in iterations:
      full_path = os.path.join(file+j, k, 'predict_results.json')
      #print(full_path)
      if os.path.exists(full_path):
        with open(full_path, "r") as f:
          json_data = json.load(f)
            # get the specific element from the JSON data
        result = json_data["predict_bleu"]
        dictionary[j].append(result)
      else:
        print('Incomplete',k)
        #dictionary[j].append('Incomplete'+k)

  return dictionary
```
