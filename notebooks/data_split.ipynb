## Swahili

```python
df = pd.read_csv('swahilisawa.csv')
```

```python
df.shape
```

```python
df['src_len'] = df.apply(lambda row: len(row.source_sentence.split()), axis = 1)
df['tgt_len'] = df.apply(lambda row: len(row.target_sentence.split()), axis = 1)
```

```python
df = df[df['src_len']<=100]
df = df[df['tgt_len']<=100]
```

```python
df.shape
```

```python
# drop conflicting translations
df.drop_duplicates(subset='source_sentence', inplace=True)
df.drop_duplicates(subset='target_sentence', inplace=True)
print(df.shape)
```

```python
# drop duplicate translations
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df.head()
```

```python
df = df[['source_sentence','target_sentence']]
```

```python
from sklearn.model_selection import KFold
```

```python
?KFold
```

```python
source_language = 'en'
target_language = 'sw'
```

```python
kfold = KFold(5, shuffle = True, random_state=1)
count = 0

for train, test in kfold.split(df):
  count +=1
  train = train.tolist()
  test = test.tolist()
  # print(type(df.loc[train]),type(df.loc[test]))
  # print(df.loc[train].shape,df.loc[test].shape)

  path = os.getcwd() 
  data_path = 'data/sw'

  it = 'k'+str(count)
  
  os.chdir(os.path.join(path, data_path))
  if not os.path.exists(it):
    os.mkdir(it)
  os.chdir(it)


  # Splitting train and validation set
  train = df.loc[train]
  test = df.loc[test]
  print(train.shape)
  num_valid = 1000

  dev = train.tail(num_valid) 
  stripped = train.drop(train.tail(num_valid).index)

  fin_train = stripped.head(30000)
  rem = stripped.tail(stripped.shape[0]-30000)

  fin_train.to_csv(target_language+'train.csv',index=False)
  dev.to_csv('dev.csv',index=False)
  rem.to_csv('rem.csv',index=False)
  test.to_csv('test.csv',index=False)

  print(fin_train.shape, dev.shape, rem.shape, test.shape)

  # Dev   
  with open("dev."+source_language, "w") as src_file, open("dev."+target_language, "w") as trg_file:
    for index, row in dev.iterrows():
      src_file.write(row["source_sentence"]+"\n")
      trg_file.write(row["target_sentence"]+"\n")

  # Test
  with open("test."+source_language, "w") as src_file, open("test."+target_language, "w") as trg_file:
    for index, row in test.iterrows():
      src_file.write(row["source_sentence"]+"\n")
      trg_file.write(row["target_sentence"]+"\n")

  os.chdir(path)
```

```python
df = df[df['src_len']<=100]
df = df[df['tgt_len']<=100]
```
